```python
import pandas as pd 
from apyori import apriori
```

```python
data = pd.read_csv('DiapersAndBeer.csv' , header=None )
```

```python
records = []
```

```python
for i in range(0,len(data)):
  records.append([str(data.values[i,j]) for j in range (0, len(data.columns))])
```

```python
rules = apriori(records , min_length=2)
results= list(rules)
for item in results: 
  print()
  print(item)
  print()
  print("------------------------")
```

```python
import pandas as pd 
data1 = pd.read_csv("Groceries.csv" , header=None)
records1=[]
for i in range (0,len(data1)):
  records1.append([str(data1.values[i,j]) for j in range (0, len(data1.columns))])
```

```python
rules = apriori(records1 , min_support=0.025 , min_length=2, min_lift=1.1)
```

```python
results = list(rules)
```

```python
for item in results:
  if not 'nan' in str(item):
    print()
    print(item)
    print()
    print("------------------------------")
```

```python
data2 = pd.read_csv("Auto-MPG.csv")
print(data2.describe())
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
coefs = np.corrcoef(data2['mpg'], data2['weight'])
plt.scatter(data2['mpg'],data2['weight'])
plt.title('MPG and Weight Correlation: '+str(coefs[0,1]))
plt.show()
```
